```python
# ==============================
# MobileNetV2 + Metadata (Early Fusion with ATTENTION)
# ==============================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.layers import (Input, Dense, Dropout, GlobalAveragePooling2D,
                                     Concatenate, Embedding, Flatten, BatchNormalization,
                                     Multiply) # <-- Added Multiply
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.applications import MobileNetV2
from tensorflow.python.framework.convert_to_constants import convert_variables_to_constants_v2

# ==============================
# PATHS (adjust)
# ==============================
# !! IMPORTANT: Update these paths to match your local environment or Kaggle/Colab setup !!
DATA_PATH = "/kaggle/input/paddy-leaf-small/Small-80"
TRAIN_DIR = os.path.join(DATA_PATH, "train")
TEST_DIR  = os.path.join(DATA_PATH, "test")
METADATA_PATH = "/kaggle/input/small-meatdata/metadata.csv"

# Check if we are in a local environment (not Kaggle) and adjust paths
if 'kaggle' not in os.getcwd():
    print("Not on Kaggle, adjusting paths for local environment...")
    # --- EDIT THESE 3 PATHS FOR YOUR LOCAL MACHINE ---
    DATA_PATH = "path/to/your/images" # e.g., "C:/Users/YourUser/Desktop/Paddy_Data"
    METADATA_PATH = "metadata.csv"    # e.g., "C:/Users/YourUser/Desktop/Paddy_Data/metadata.csv"
    # ---------------------------------------------------
    
    # Re-join train/test dirs based on new DATA_PATH
    TRAIN_DIR = os.path.join(DATA_PATH, "train")
    TEST_DIR = os.path.join(DATA_PATH, "test")

IMG_SIZE = 224
BATCH_SIZE = 32
SEED = 42
EPOCHS = 150
LR = 1e-4

# ==============================
# LOAD METADATA
# ==============================
try:
    df = pd.read_csv(METADATA_PATH)
except FileNotFoundError:
    print(f"Error: Metadata file not found at {METADATA_PATH}")
    print("Please update METADATA_PATH in the script.")
    exit()

if not df['image_id'].iloc[0].lower().endswith('.jpg'):
    df['image_id'] = df['image_id'].astype(str) + ".jpg"

df['image_path'] = df.apply(
    lambda r: os.path.join(TRAIN_DIR if r['split']=='train' else TEST_DIR, r['label'], r['image_id']),
    axis=1
)
df = df[df['image_path'].apply(os.path.exists)].reset_index(drop=True)

if len(df) == 0:
    print("Error: No valid image paths were found.")
    print(f"Please check that DATA_PATH ('{DATA_PATH}') and METADATA_PATH ('{METADATA_PATH}') are correct.")
    exit()

# Encode label + variety
label_enc = LabelEncoder()
df['label_enc'] = label_enc.fit_transform(df['label'])
num_classes = df['label_enc'].nunique()

var_enc = LabelEncoder()
df['variety_enc'] = var_enc.fit_transform(df['variety'])
num_varieties = df['variety_enc'].nunique()

df['age_norm'] = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min())

print(f"âœ… Dataset loaded: {len(df)} samples, {num_classes} classes")

# ==============================
# SPLITS
# ==============================
train_df_full = df[df['split']=='train']
test_df = df[df['split']=='test']

train_df, val_df = train_test_split(
    train_df_full,
    test_size=0.25,
    stratify=train_df_full['label_enc'],
    random_state=SEED
)

print(f"Train={len(train_df)}, Val={len(val_df)}, Test={len(test_df)}")

# ==============================
# DATASET CREATOR
# ==============================
def create_dataset(df_local, batch_size=BATCH_SIZE, img_size=(IMG_SIZE, IMG_SIZE), shuffle=True):
    paths = df_local['image_path'].values.astype(str)
    varieties = df_local['variety_enc'].values.astype(np.int32)
    ages = df_local['age_norm'].values.astype(np.float32)
    labels = to_categorical(df_local['label_enc'].values, num_classes=num_classes).astype(np.float32)

    ds = tf.data.Dataset.from_tensor_slices((paths, varieties, ages, labels))
    if shuffle:
        ds = ds.shuffle(len(df_local), seed=SEED)

    def _load_fn(path, var, age, label):
        img = tf.io.read_file(path)
        img = tf.io.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, img_size)
        img = tf.cast(img, tf.float32) / 255.0
        return (img, var, age), label

    return ds.map(_load_fn, num_parallel_calls=tf.data.AUTOTUNE)\
             .batch(batch_size)\
             .prefetch(tf.data.AUTOTUNE)

train_ds = create_dataset(train_df)
val_ds   = create_dataset(val_df, shuffle=False)
test_ds  = create_dataset(test_df, shuffle=False)

# ==============================
# MODEL: MobileNetV2 + Metadata + ATTENTION
# --- THIS FUNCTION IS UPDATED ---
# ==============================
def create_mobilenetv2_metadata(img_size=IMG_SIZE, num_classes=num_classes, num_varieties=num_varieties):
    # Image input
    img_input = Input(shape=(img_size, img_size, 3), name='image_input')
    base = MobileNetV2(include_top=False, weights="imagenet", input_tensor=img_input)
    base.trainable = False

    x = base.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    img_feat = Dense(128, activation='relu')(x)

    # Metadata inputs
    var_in = Input(shape=(1,), name='var_input', dtype='int32')
    age_in = Input(shape=(1,), name='age_input', dtype='float32')

    var_emb = Embedding(num_varieties, 8)(var_in)
    var_emb = Flatten()(var_emb)
    var_emb = Dense(32, activation='relu')(var_emb)

    age_dense = Dense(16, activation='relu')(age_in)

    # Fusion
    fused = Concatenate()([img_feat, var_emb, age_dense])

    # <<< START: NEW ATTENTION BLOCK (SQUEEZE-AND-EXCITE) >>>
    fused_channels = int(fused.shape[-1])       # 128 + 32 + 16 = 176 features
    attention_bottleneck = fused_channels // 8  # 176 // 8 = 22
    
    # 1. Squeeze
    se = Dense(attention_bottleneck, activation='relu', name='attn_squeeze')(fused)
    
    # 2. Excite
    attention_weights = Dense(fused_channels, activation='sigmoid', name='attn_excite_weights')(se)
    
    # 3. Apply
    fused_with_attention = Multiply(name='attn_apply')([fused, attention_weights])
    # <<< END: NEW ATTENTION BLOCK >>>

    # Classification Head (uses attended features)
    f = Dense(256, activation='relu')(fused_with_attention) # <-- Use new vector
    f = BatchNormalization()(f)
    f = Dropout(0.4)(f)
    f = Dense(128, activation='relu')(f)
    f = Dropout(0.3)(f)
    f = Dense(64, activation='relu')(f)
    f = Dropout(0.2)(f)

    out = Dense(num_classes, activation='softmax')(f)

    return Model(inputs=[img_input, var_in, age_in], outputs=out, name="MobileNetV2_Attention_Metadata")

model = create_mobilenetv2_metadata()
model.compile(optimizer=Adam(LR), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ==============================
# CALLBACKS
# ==============================
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1),
    ModelCheckpoint("mobilenetv2_attention_metadata_best.h5", # <-- Updated name
                    save_best_only=True, 
                    monitor='val_accuracy', 
                    verbose=1)
]

# ==============================
# TRAIN
# ==============================
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=callbacks, verbose=1)
model.save("mobilenetv2_attention_metadata_final.h5") # <-- Updated name

# ==============================
# PLOTS
# ==============================
plt.figure(figsize=(10,4))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend(); plt.title("Accuracy (MobileNetV2 + Attention)")
plt.savefig("mobilenetv2_attention_metadata_acc.png") # <-- Updated name

plt.figure(figsize=(10,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend(); plt.title("Loss (MobileNetV2 + Attention)")
plt.savefig("mobilenetv2_attention_metadata_loss.png") # <-- Updated name

# ==============================
# TEST EVAL
# ==============================
test_loss, test_acc = model.evaluate(test_ds)
print(f"\nâœ… Test accuracy: {test_acc*100:.2f}% | loss: {test_loss:.4f}")

y_true, y_pred = [], []
for (imgs, vars_, ages), labs in test_ds:
    preds = model.predict([imgs, vars_, ages], verbose=0)
    y_true.extend(np.argmax(labs.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=label_enc.classes_))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_enc.classes_, yticklabels=label_enc.classes_)
plt.title("Confusion Matrix - MobileNetV2 + Attention Metadata") # <-- Updated title
plt.xlabel("Predicted"); plt.ylabel("True")
plt.savefig("mobilenetv2_attention_metadata_cm.png") # <-- Updated name

# ==============================
# GFLOPS + PARAMS
# ==============================
def print_model_gflops(model, img_size=224):
    try:
        concrete_func = tf.function(lambda x: model([x, tf.zeros((1,1), tf.int32), tf.zeros((1,1), tf.float32)], training=False))\
                            .get_concrete_function(tf.TensorSpec([1,img_size,img_size,3], tf.float32))
        frozen_func = convert_variables_to_constants_v2(concrete_func)
        graph_def = frozen_func.graph.as_graph_def()
        with tf.Graph().as_default() as graph:
            tf.import_graph_def(graph_def, name='')
            run_meta = tf.compat.v1.RunMetadata()
            opts = tf.compat.v1.profiler.ProfileOptionBuilder.float_operation()
            flops = tf.compat.v1.profiler.profile(graph=graph, run_meta=run_meta, cmd='op', options=opts)
            gflops = flops.total_float_ops/1e9 if flops else 0
            print(f"\nâš¡ GFLOPS: {gflops:.3f}")
    except Exception as e:
        print("âš ï¸ GFLOPS calc failed:", e)
    print(f"ğŸ“¦ Total Parameters: {model.count_params():,}")

print_model_gflops(model, img_size=IMG_SIZE)

print("\nâœ… Done.")
```

```python
# ===============================================================
# MobileNetV2 + CBAM (Channel + Spatial Attention) + Metadata (Early Fusion)
# ===============================================================

import os, warnings
warnings.filterwarnings('ignore')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf
from tensorflow.keras.layers import (Input, Dense, Dropout, GlobalAveragePooling2D, GlobalMaxPooling2D,
                                     Concatenate, Embedding, Flatten, BatchNormalization, Multiply, Add,
                                     Activation, Conv2D, Reshape, Lambda)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import backend as K

# ===============================================================
# CONFIG
# ===============================================================
DATA_PATH = "/kaggle/input/paddy-leaf-small/Small-80"
TRAIN_DIR = os.path.join(DATA_PATH, "train")
TEST_DIR  = os.path.join(DATA_PATH, "test")
METADATA_PATH = "/kaggle/input/small-meatdata/metadata.csv"

IMG_SIZE = 224
BATCH_SIZE = 32
SEED = 42
EPOCHS = 150
LR = 1e-4

# ===============================================================
# LOAD METADATA
# ===============================================================
df = pd.read_csv(METADATA_PATH)
if not df['image_id'].iloc[0].lower().endswith('.jpg'):
    df['image_id'] = df['image_id'].astype(str) + ".jpg"

df['image_path'] = df.apply(
    lambda r: os.path.join(TRAIN_DIR if r['split']=='train' else TEST_DIR, r['label'], r['image_id']),
    axis=1
)
df = df[df['image_path'].apply(os.path.exists)].reset_index(drop=True)

# Encode label + variety
label_enc = LabelEncoder()
df['label_enc'] = label_enc.fit_transform(df['label'])
num_classes = df['label_enc'].nunique()

var_enc = LabelEncoder()
df['variety_enc'] = var_enc.fit_transform(df['variety'])
num_varieties = df['variety_enc'].nunique()

df['age_norm'] = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min())

print(f"âœ… Dataset loaded: {len(df)} samples, {num_classes} classes")

# ===============================================================
# SPLITS
# ===============================================================
train_df_full = df[df['split']=='train']
test_df = df[df['split']=='test']

train_df, val_df = train_test_split(
    train_df_full,
    test_size=0.25,
    stratify=train_df_full['label_enc'],
    random_state=SEED
)

print(f"Train={len(train_df)}, Val={len(val_df)}, Test={len(test_df)}")

# ===============================================================
# DATASET CREATOR
# ===============================================================
def create_dataset(df_local, batch_size=BATCH_SIZE, img_size=(IMG_SIZE, IMG_SIZE), shuffle=True):
    paths = df_local['image_path'].values.astype(str)
    varieties = df_local['variety_enc'].values.astype(np.int32)
    ages = df_local['age_norm'].values.astype(np.float32)
    labels = to_categorical(df_local['label_enc'].values, num_classes=num_classes).astype(np.float32)

    ds = tf.data.Dataset.from_tensor_slices((paths, varieties, ages, labels))
    if shuffle:
        ds = ds.shuffle(len(df_local), seed=SEED)

    def _load_fn(path, var, age, label):
        img = tf.io.read_file(path)
        img = tf.io.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, img_size)
        img = tf.cast(img, tf.float32) / 255.0
        return (img, var, age), label

    return ds.map(_load_fn, num_parallel_calls=tf.data.AUTOTUNE)\
             .batch(batch_size)\
             .prefetch(tf.data.AUTOTUNE)

train_ds = create_dataset(train_df)
val_ds   = create_dataset(val_df, shuffle=False)
test_ds  = create_dataset(test_df, shuffle=False)

# ===============================================================
# CBAM Attention Block
# ===============================================================
def cbam_block(feature_map, ratio=8, name_prefix="cbam"):
    ch = feature_map.shape[-1]

    # ----- Channel Attention -----
    shared_dense1 = Dense(ch // ratio, activation='relu', name=f"{name_prefix}_ch_dense1")
    shared_dense2 = Dense(ch, activation=None, name=f"{name_prefix}_ch_dense2")
    avg_pool = shared_dense2(shared_dense1(GlobalAveragePooling2D()(feature_map)))
    max_pool = shared_dense2(shared_dense1(GlobalMaxPooling2D()(feature_map)))
    ch_attn = Activation('sigmoid')(Add()([avg_pool, max_pool]))
    ch_attn = Reshape((1, 1, ch))(ch_attn)
    x = Multiply()([feature_map, ch_attn])

    # ----- Spatial Attention -----
    avg_sp = Lambda(lambda t: K.mean(t, axis=-1, keepdims=True))(x)
    max_sp = Lambda(lambda t: K.max(t, axis=-1, keepdims=True))(x)
    concat = Concatenate(axis=-1)([avg_sp, max_sp])
    sa = Conv2D(1, kernel_size=7, padding='same', activation='sigmoid', name=f"{name_prefix}_sp_conv")(concat)
    x = Multiply()([x, sa])
    return x

# ===============================================================
# MODEL: MobileNetV2 + CBAM + Metadata
# ===============================================================
def create_mobilenetv2_cbam_metadata(img_size=IMG_SIZE, num_classes=num_classes, num_varieties=num_varieties):
    # Image input
    img_input = Input(shape=(img_size, img_size, 3), name='image_input')
    base = MobileNetV2(include_top=False, weights="imagenet", input_tensor=img_input)
    base.trainable = False

    # Extract features
    x = base.output
    x = cbam_block(x, ratio=8, name_prefix="cbam_mnv2")
    x = GlobalAveragePooling2D()(x)

    # Image dense head
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    img_feat = Dense(128, activation='relu')(x)

    # Metadata inputs
    var_in = Input(shape=(1,), name='var_input', dtype='int32')
    age_in = Input(shape=(1,), name='age_input', dtype='float32')

    var_emb = Embedding(num_varieties, 8)(var_in)
    var_emb = Flatten()(var_emb)
    var_emb = Dense(32, activation='relu')(var_emb)

    age_dense = Dense(16, activation='relu')(age_in)

    # Fusion
    fused = Concatenate()([img_feat, var_emb, age_dense])
    f = Dense(256, activation='relu')(fused)
    f = BatchNormalization()(f)
    f = Dropout(0.4)(f)
    f = Dense(128, activation='relu')(f)
    f = Dropout(0.3)(f)
    f = Dense(64, activation='relu')(f)
    f = Dropout(0.2)(f)

    out = Dense(num_classes, activation='softmax')(f)
    return Model(inputs=[img_input, var_in, age_in], outputs=out, name="MobileNetV2_CBAM_Metadata")

model = create_mobilenetv2_cbam_metadata()
model.compile(optimizer=Adam(LR), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ===============================================================
# CALLBACKS
# ===============================================================
callbacks = [
    EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=6, verbose=1),
    ModelCheckpoint("mobilenetv2_cbam_best.h5", monitor='val_accuracy', save_best_only=True, verbose=1)
]

# ===============================================================
# TRAIN
# ===============================================================
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=callbacks, verbose=1)
model.save("mobilenetv2_cbam_final.h5")

# ===============================================================
# TRAINING PLOTS
# ===============================================================
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("MobileNetV2 + CBAM - Accuracy")
plt.xlabel("Epochs"); plt.ylabel("Accuracy")
plt.legend(); plt.grid(alpha=0.3)
plt.savefig("mobilenetv2_cbam_acc.png", dpi=150)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("MobileNetV2 + CBAM - Loss")
plt.xlabel("Epochs"); plt.ylabel("Loss")
plt.legend(); plt.grid(alpha=0.3)
plt.savefig("mobilenetv2_cbam_loss.png", dpi=150)
plt.show()

# ===============================================================
# TEST EVALUATION
# ===============================================================
test_loss, test_acc = model.evaluate(test_ds)
print(f"\nâœ… Test accuracy: {test_acc*100:.2f}% | loss: {test_loss:.4f}")

y_true, y_pred = [], []
for (imgs, vars_, ages), labs in test_ds:
    preds = model.predict([imgs, vars_, ages], verbose=0)
    y_true.extend(np.argmax(labs.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=label_enc.classes_))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_enc.classes_, yticklabels=label_enc.classes_)
plt.title("Confusion Matrix - MobileNetV2 + CBAM + Metadata")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout()
plt.savefig("mobilenetv2_cbam_confusion_matrix.png", dpi=150)
plt.show()

print(f"\nğŸ“¦ Total Parameters: {model.count_params():,}")
print("\nâœ… Done: MobileNetV2 + CBAM + Metadata training complete.")
```
