```python
# DenseNet201 + Metadata (Early Fusion with ATTENTION)
# ====================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.layers import (Input, Dense, Dropout, GlobalAveragePooling2D,
                                     Concatenate, Embedding, Flatten, BatchNormalization,
                                     Multiply) # <-- Added Multiply
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

# --------------------
# Paths & params â€” EDIT if needed
# --------------------
# !! IMPORTANT: Update these paths to match your local environment or Kaggle/Colab setup !!
DATA_PATH = "/kaggle/input/paddy-leaf-small/Small-80"
TRAIN_DIR = os.path.join(DATA_PATH, "train")
TEST_DIR = os.path.join(DATA_PATH, "test")
METADATA_PATH = "/kaggle/input/small-meatdata/metadata.csv" # update to your CSV path
PRETRAINED_WEIGHTS = "/kaggle/input/densenet201-weights/densenet201_weights_tf_dim_ordering_tf_kernels_notop (1).h5" # update to your weights path

# Check if we are in a local environment (not Kaggle) and adjust paths
if 'kaggle' not in os.getcwd():
    print("Not on Kaggle, adjusting paths for local environment...")
    # --- EDIT THESE 3 PATHS FOR YOUR LOCAL MACHINE ---
    DATA_PATH = "path/to/your/images" # e.g., "C:/Users/YourUser/Desktop/Paddy_Data"
    METADATA_PATH = "metadata.csv"    # e.g., "C:/Users/YourUser/Desktop/Paddy_Data/metadata.csv"
    PRETRAINED_WEIGHTS = "densenet201_weights_tf_dim_ordering_tf_kernels_notop.h5" # e.g., "C:/Users/YourUser/Desktop/weights/densenet201_weights.h5"
    # ---------------------------------------------------
    
    # Re-join train/test dirs based on new DATA_PATH
    TRAIN_DIR = os.path.join(DATA_PATH, "train")
    TEST_DIR = os.path.join(DATA_PATH, "test")
    # Use 'imagenet' weights if a local path isn't found
    if not os.path.exists(PRETRAINED_WEIGHTS):
        print(f"Local weights file not found at {PRETRAINED_WEIGHTS}. Using 'imagenet' weights.")
        PRETRAINED_WEIGHTS = 'imagenet'

IMG_SIZE = 224
BATCH_SIZE = 32
SEED = 42
EPOCHS_PHASE1 = 150      # you can reduce if compute is limited
LEARNING_RATE = 1e-4

# --------------------
# Load metadata & prepare file paths
# --------------------
try:
    df = pd.read_csv(METADATA_PATH)
except FileNotFoundError:
    print(f"Error: Metadata file not found at {METADATA_PATH}")
    print("Please update METADATA_PATH in the script.")
    # In a real script, you'd exit here. For this example, we'll stop.
    exit()

# Ensure image_id includes extension. If not, append .jpg
if not df['image_id'].iloc[0].lower().endswith('.jpg'):
    df['image_id'] = df['image_id'].astype(str) + ".jpg"

# Build full path pointing to label subfolder
df['image_path'] = df.apply(
    lambda r: os.path.join(TRAIN_DIR if r['split']=='train' else TEST_DIR, r['label'], r['image_id']),
    axis=1
)

# Keep only rows whose image files exist
exists_mask = df['image_path'].apply(os.path.exists)
missing_count = (~exists_mask).sum()
if missing_count:
    print(f"âš ï¸ {missing_count} metadata rows reference missing image files â€” they will be dropped.")
df = df[exists_mask].reset_index(drop=True)

if len(df) == 0:
    print("Error: No valid image paths were found.")
    print(f"Please check that DATA_PATH ('{DATA_PATH}') and METADATA_PATH ('{METADATA_PATH}') are correct.")
    exit()

# Encode label and variety
label_encoder = LabelEncoder()
df['label_enc'] = label_encoder.fit_transform(df['label'])
num_classes = df['label_enc'].nunique()

variety_encoder = LabelEncoder()
df['variety_enc'] = variety_encoder.fit_transform(df['variety'])
num_varieties = df['variety_enc'].nunique()

# Normalize age (min-max)
df['age_norm'] = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min())

print("ğŸ“Š Dataset Info:")
print(f" â€¢ total images (after dropping missing): {len(df)}")
print(f" â€¢ classes: {num_classes} -> {list(label_encoder.classes_)}")
print(f" â€¢ varieties: {num_varieties}")

# --------------------
# Train/Val/Test split (60:20:20) using supplied split column
# --------------------
train_df_full = df[df['split']=='train']
test_df = df[df['split']=='test']

train_df, val_df = train_test_split(
    train_df_full,
    test_size=0.25,          # from train: 0.25 -> 60:20
    stratify=train_df_full['label_enc'],
    random_state=SEED
)

print("\nğŸ“Š Split sizes:")
print(f" â€¢ train: {len(train_df)}")
print(f" â€¢ val:   {len(val_df)}")
print(f" â€¢ test:  {len(test_df)}")

# --------------------
# Create tf.data datasets (image + metadata)
# --------------------
def create_dataset(df_local, batch_size=BATCH_SIZE, img_size=(IMG_SIZE, IMG_SIZE), shuffle=True):
    image_paths = df_local['image_path'].values.astype(str)
    varieties = df_local['variety_enc'].values.astype(np.int32)
    ages = df_local['age_norm'].values.astype(np.float32)
    labels = to_categorical(df_local['label_enc'].values, num_classes=num_classes).astype(np.float32)
    
    ds = tf.data.Dataset.from_tensor_slices((image_paths, varieties, ages, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(df_local), seed=SEED)
    
    def _load_fn(image_path, variety, age, label):
        # image_path: tf.string
        img = tf.io.read_file(image_path)
        # decode_jpeg is deterministic and gives channel dimension
        img = tf.io.decode_jpeg(img, channels=3)
        img.set_shape([None, None, 3])   # allow resize
        img = tf.image.resize(img, img_size)
        img = tf.cast(img, tf.float32) / 255.0
        # ensure correct dtypes for metadata
        variety = tf.cast(variety, tf.int32)      # for Embedding input
        age = tf.cast(age, tf.float32)
        return (img, variety, age), label
    
    ds = ds.map(_load_fn, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = create_dataset(train_df, batch_size=BATCH_SIZE, shuffle=True)
val_ds   = create_dataset(val_df,   batch_size=BATCH_SIZE, shuffle=False)
test_ds  = create_dataset(test_df,  batch_size=BATCH_SIZE, shuffle=False)

# quick sanity check
for (img_batch, var_batch, age_batch), lab_batch in train_ds.take(1):
    print(f"\nSanity check batch shapes: img {img_batch.shape}, var {var_batch.shape}, age {age_batch.shape}, label {lab_batch.shape}")
    break

# --------------------
# Build DenseNet201 (Early Fusion with ATTENTION)
# --- THIS FUNCTION IS UPDATED ---
# --------------------
def create_densenet_metadata_model(img_size=IMG_SIZE, num_classes=num_classes, num_varieties=num_varieties):
    # Image input
    img_input = Input(shape=(img_size, img_size, 3), name='image_input')
    # DenseNet201 backbone â€” use provided weights path or 'imagenet'
    base = DenseNet201(include_top=False, weights=PRETRAINED_WEIGHTS, input_tensor=img_input)
    base.trainable = False  # no fine-tuning
    
    x = base.output
    x = GlobalAveragePooling2D(name='gap')(x)
    x = Dense(512, activation='relu', name='img_dense_1')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu', name='img_dense_2')(x)
    x = BatchNormalization()(x)
    # image_features final
    image_features = Dense(128, activation='relu', name='img_features')(x)
    
    # Metadata branch
    var_input = Input(shape=(1,), name='var_input', dtype='int32')
    age_input = Input(shape=(1,), name='age_input', dtype='float32')
    
    var_emb = Embedding(input_dim=num_varieties, output_dim=8, name='var_emb')(var_input)
    var_emb = Flatten()(var_emb)
    var_emb = Dense(32, activation='relu', name='var_dense')(var_emb)
    
    age_dense = Dense(16, activation='relu', name='age_dense')(age_input)
    
    # Early fusion: concat image features + metadata features
    fused = Concatenate(name='fusion')([image_features, var_emb, age_dense])
    
    # <<< START: NEW ATTENTION BLOCK (SQUEEZE-AND-EXCITE) >>>
    # This block learns to weigh the importance of each feature in the fused vector.
    # It allows the model to "pay more attention" to metadata if the image is ambiguous.
    
    # Dynamically get the number of channels (features) from the fused layer
    fused_channels = int(fused.shape[-1])  # This will be 128 + 32 + 16 = 176
    
    # Set a reduction ratio for the "Squeeze" bottleneck
    attention_bottleneck = fused_channels // 8  # e.g., 176 // 8 = 22
    
    # 1. Squeeze: Learn a compact summary of all features
    se = Dense(attention_bottleneck, activation='relu', name='attn_squeeze')(fused)
    
    # 2. Excite: Generate a per-feature weight (0 to 1) from that summary
    attention_weights = Dense(fused_channels, activation='sigmoid', name='attn_excite_weights')(se)
    
    # 3. Apply: Multiply the original features by their learned weights
    # The model now learns *which features are important* for each prediction
    fused_with_attention = Multiply(name='attn_apply')([fused, attention_weights])
    # <<< END: NEW ATTENTION BLOCK >>>
    
    
    # Classification head (at least 3 dense layers)
    # We now feed the *attended* features into the classifier
    f = Dense(256, activation='relu', name='fusion_dense_1')(fused_with_attention) # <-- Use new vector
    f = BatchNormalization()(f)
    f = Dropout(0.4)(f)
    
    f = Dense(128, activation='relu', name='fusion_dense_2')(f)
    f = BatchNormalization()(f)
    f = Dropout(0.3)(f)
    
    f = Dense(64, activation='relu', name='fusion_dense_3')(f)
    f = Dropout(0.2)(f)
    
    out = Dense(num_classes, activation='softmax', name='predictions')(f)
    
    model = Model(inputs=[img_input, var_input, age_input], outputs=out, name='DenseNet201_AttentionFusion_Metadata')
    return model, base

model, base_model = create_densenet_metadata_model()
model.compile(optimizer=Adam(LEARNING_RATE), loss='categorical_crossentropy', metrics=['accuracy'])
# model.summary()

# --------------------
# Callbacks: EarlyStopping + ReduceLROnPlateau + ModelCheckpoint
# --------------------
checkpoint_path = "densenet_attention_metadata_best.h5"
callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=6, verbose=1, min_lr=1e-7),
    EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True, verbose=1),
    ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1)
]

# --------------------
# Train (no fine-tuning)
# --------------------
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_PHASE1,
    callbacks=callbacks,
    verbose=1
)

# Save the final model (best weights already restored by EarlyStopping if used)
model.save("densenet_attention_metadata_final.h5")
print("âœ… Saved final model: densenet_attention_metadata_final.h5")

# --------------------
# Plot & Save training curves
# --------------------
def save_training_plots(history, prefix="densenet_attention_metadata"):
    # Accuracy
    plt.figure(figsize=(10,4))
    plt.plot(history.history['accuracy'], label='train_acc')
    plt.plot(history.history['val_accuracy'], label='val_acc')
    plt.title('Accuracy')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.legend()
    plt.grid(alpha=0.3)
    acc_png = f"{prefix}_accuracy.png"
    plt.savefig(acc_png, bbox_inches='tight', dpi=150)
    print(f"Saved {acc_png}")
    plt.close()
    
    # Loss
    plt.figure(figsize=(10,4))
    plt.plot(history.history['loss'], label='train_loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.title('Loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend()
    plt.grid(alpha=0.3)
    loss_png = f"{prefix}_loss.png"
    plt.savefig(loss_png, bbox_inches='tight', dpi=150)
    print(f"Saved {loss_png}")
    plt.close()

save_training_plots(history, prefix="densenet_attention_metadata")

# --------------------
# Evaluation on test set
# --------------------
test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print(f"\nâœ… Test accuracy: {test_acc*100:.2f}% | test loss: {test_loss:.4f}")

# Gather predictions for detailed metrics
y_true = []
y_pred = []

for (imgs, vars_, ages), labs in test_ds:
    preds = model.predict([imgs, vars_, ages], verbose=0)
    y_true.extend(np.argmax(labs.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

# Classification report & confusion matrix
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix - DenseNet201 + Attention Metadata")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.savefig("densenet_attention_metadata_confusion_matrix.png", bbox_inches='tight', dpi=150)
print("Saved densenet_attention_metadata_confusion_matrix.png")
plt.show()

# Weighted metrics
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
print(f"\nWeighted metrics: Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}")

# --------------------
# Print model params
# --------------------
print(f"\nTotal parameters: {model.count_params():,}")
# print("Layers and number of weight arrays:")
# for layer in model.layers:
#     print(f" - {layer.name}: {len(layer.get_weights())} weight tensors")

# Done
print("\nâœ… Done. Model, plots and confusion matrix saved in the working directory.")
```

```python
# DenseNet201 + Metadata (Early Fusion with ATTENTION + FINE-TUNING)
# =================================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.layers import (Input, Dense, Dropout, GlobalAveragePooling2D,
                                     Concatenate, Embedding, Flatten, BatchNormalization,
                                     Multiply, RandomFlip, RandomRotation, RandomBrightness, RandomZoom)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

# --------------------
# Paths & params â€” EDIT if needed
# --------------------
# !! IMPORTANT: Update these paths to match your local environment or Kaggle/Colab setup !!
DATA_PATH = "/kaggle/input/paddy-leaf-small/Small-80"
TRAIN_DIR = os.path.join(DATA_PATH, "train")
TEST_DIR = os.path.join(DATA_PATH, "test")
METADATA_PATH = "/kaggle/input/small-meatdata/metadata.csv" # update to your CSV path
PRETRAINED_WEIGHTS = "/kaggle/input/densenet201-weights/densenet201_weights_tf_dim_ordering_tf_kernels_notop (1).h5" # update to your weights path

# Check if we are in a local environment (not Kaggle) and adjust paths
if 'kaggle' not in os.getcwd():
    print("Not on Kaggle, adjusting paths for local environment...")
    # --- EDIT THESE 3 PATHS FOR YOUR LOCAL MACHINE ---
    DATA_PATH = "path/to/your/images" # e.g., "C:/Users/YourUser/Desktop/Paddy_Data"
    METADATA_PATH = "metadata.csv"    # e.g., "C:/Users/YourUser/Desktop/Paddy_Data/metadata.csv"
    PRETRAINED_WEIGHTS = "densenet201_weights_tf_dim_ordering_tf_kernels_notop.h5" # e.g., "C:/Users/YourUser/Desktop/weights/densenet201_weights.h5"
    # ---------------------------------------------------
    
    # Re-join train/test dirs based on new DATA_PATH
    TRAIN_DIR = os.path.join(DATA_PATH, "train")
    TEST_DIR = os.path.join(DATA_PATH, "test")
    # Use 'imagenet' weights if a local path isn't found
    if not os.path.exists(PRETRAINED_WEIGHTS):
        print(f"Local weights file not found at {PRETRAINED_WEIGHTS}. Using 'imagenet' weights.")
        PRETRAINED_WEIGHTS = 'imagenet'

IMG_SIZE = 224
BATCH_SIZE = 32
SEED = 42
EPOCHS_PHASE1 = 60      # Epochs for the initial head training
EPOCHS_PHASE2 = 25      # Epochs for fine-tuning
LEARNING_RATE = 1e-4

# --------------------
# Load metadata & prepare file paths
# --------------------
try:
    df = pd.read_csv(METADATA_PATH)
except FileNotFoundError:
    print(f"Error: Metadata file not found at {METADATA_PATH}")
    print("Please update METADATA_PATH in the script.")
    exit()

# (The rest of the metadata loading is the same...)
if not df['image_id'].iloc[0].lower().endswith('.jpg'):
    df['image_id'] = df['image_id'].astype(str) + ".jpg"

df['image_path'] = df.apply(
    lambda r: os.path.join(TRAIN_DIR if r['split']=='train' else TEST_DIR, r['label'], r['image_id']),
    axis=1
)
exists_mask = df['image_path'].apply(os.path.exists)
missing_count = (~exists_mask).sum()
if missing_count:
    print(f"âš ï¸ {missing_count} metadata rows reference missing image files â€” they will be dropped.")
df = df[exists_mask].reset_index(drop=True)

if len(df) == 0:
    print("Error: No valid image paths were found.")
    print(f"Please check that DATA_PATH ('{DATA_PATH}') and METADATA_PATH ('{METADATA_PATH}') are correct.")
    exit()

# Encode label and variety
label_encoder = LabelEncoder()
df['label_enc'] = label_encoder.fit_transform(df['label'])
num_classes = df['label_enc'].nunique()

variety_encoder = LabelEncoder()
df['variety_enc'] = variety_encoder.fit_transform(df['variety'])
num_varieties = df['variety_enc'].nunique()

# Normalize age (min-max)
df['age_norm'] = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min())

print("ğŸ“Š Dataset Info:")
print(f" â€¢ total images (after dropping missing): {len(df)}")
print(f" â€¢ classes: {num_classes} -> {list(label_encoder.classes_)}")
print(f" â€¢ varieties: {num_varieties}")

# --------------------
# Train/Val/Test split
# --------------------
train_df_full = df[df['split']=='train']
test_df = df[df['split']=='test']

train_df, val_df = train_test_split(
    train_df_full,
    test_size=0.25,          # from train: 0.25 -> 60:20
    stratify=train_df_full['label_enc'],
    random_state=SEED
)

print("\nğŸ“Š Split sizes:")
print(f" â€¢ train: {len(train_df)}")
print(f" â€¢ val:   {len(val_df)}")
print(f" â€¢ test:  {len(test_df)}")

# --------------------
# <<< NEW: Data Augmentation Layers >>>
# --------------------
data_augmentation = tf.keras.Sequential([
    RandomFlip("horizontal", seed=SEED),
    RandomRotation(0.1, seed=SEED),
    RandomBrightness(0.1, seed=SEED),
    RandomZoom(0.1, seed=SEED),
], name="data_augmentation")

# --------------------
# Create tf.data datasets (with Augmentation)
# --------------------
def create_dataset(df_local, batch_size=BATCH_SIZE, img_size=(IMG_SIZE, IMG_SIZE), shuffle=True, augment=False):
    image_paths = df_local['image_path'].values.astype(str)
    varieties = df_local['variety_enc'].values.astype(np.int32)
    ages = df_local['age_norm'].values.astype(np.float32)
    labels = to_categorical(df_local['label_enc'].values, num_classes=num_classes).astype(np.float32)
    
    ds = tf.data.Dataset.from_tensor_slices((image_paths, varieties, ages, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(df_local), seed=SEED)
    
    def _load_fn(image_path, variety, age, label):
        img = tf.io.read_file(image_path)
        img = tf.io.decode_jpeg(img, channels=3)
        img.set_shape([None, None, 3])   # allow resize
        img = tf.image.resize(img, img_size)
        img = tf.cast(img, tf.float32) / 255.0
        variety = tf.cast(variety, tf.int32)
        age = tf.cast(age, tf.float32)
        return (img, variety, age), label
    
    ds = ds.map(_load_fn, num_parallel_calls=tf.data.AUTOTUNE)
    
    # <<< NEW: Apply augmentation ONLY if augment=True >>>
    if augment:
        # We un-tuple, apply augmentation to the image, then re-tuple
        ds = ds.map(lambda inputs, label: ((data_augmentation(inputs[0], training=True), inputs[1], inputs[2]), label),
                    num_parallel_calls=tf.data.AUTOTUNE)
    
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

# --- Apply augmentation ONLY to train_ds ---
train_ds = create_dataset(train_df, batch_size=BATCH_SIZE, shuffle=True, augment=True) # <-- augment=True
val_ds   = create_dataset(val_df,   batch_size=BATCH_SIZE, shuffle=False, augment=False)
test_ds  = create_dataset(test_df,  batch_size=BATCH_SIZE, shuffle=False, augment=False)

# Sanity check
for (img_batch, var_batch, age_batch), lab_batch in train_ds.take(1):
    print(f"\nSanity check batch shapes: img {img_batch.shape}, var {var_batch.shape}, age {age_batch.shape}, label {lab_batch.shape}")
    break

# --------------------
# Build DenseNet201 (Early Fusion with ATTENTION)
# --------------------
def create_densenet_metadata_model(img_size=IMG_SIZE, num_classes=num_classes, num_varieties=num_varieties):
    # Image input
    img_input = Input(shape=(img_size, img_size, 3), name='image_input')
    # DenseNet201 backbone
    base = DenseNet201(include_top=False, weights=PRETRAINED_WEIGHTS, input_tensor=img_input)
    # --- We set trainable to False for PHASE 1 ---
    base.trainable = False
    
    x = base.output
    x = GlobalAveragePooling2D(name='gap')(x)
    x = Dense(512, activation='relu', name='img_dense_1')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu', name='img_dense_2')(x)
    x = BatchNormalization()(x)
    image_features = Dense(128, activation='relu', name='img_features')(x)
    
    # Metadata branch
    var_input = Input(shape=(1,), name='var_input', dtype='int32')
    age_input = Input(shape=(1,), name='age_input', dtype='float32')
    
    var_emb = Embedding(input_dim=num_varieties, output_dim=8, name='var_emb')(var_input)
    var_emb = Flatten()(var_emb)
    var_emb = Dense(32, activation='relu', name='var_dense')(var_emb)
    
    age_dense = Dense(16, activation='relu', name='age_dense')(age_input)
    
    # Early fusion
    fused = Concatenate(name='fusion')([image_features, var_emb, age_dense])
    
    # Attention Block
    fused_channels = int(fused.shape[-1])
    attention_bottleneck = fused_channels // 8
    se = Dense(attention_bottleneck, activation='relu', name='attn_squeeze')(fused)
    attention_weights = Dense(fused_channels, activation='sigmoid', name='attn_excite_weights')(se)
    fused_with_attention = Multiply(name='attn_apply')([fused, attention_weights])
    
    # Classification head
    f = Dense(256, activation='relu', name='fusion_dense_1')(fused_with_attention)
    f = BatchNormalization()(f)
    f = Dropout(0.4)(f)
    f = Dense(128, activation='relu', name='fusion_dense_2')(f)
    f = BatchNormalization()(f)
    f = Dropout(0.3)(f)
    f = Dense(64, activation='relu', name='fusion_dense_3')(f)
    f = Dropout(0.2)(f)
    
    out = Dense(num_classes, activation='softmax', name='predictions')(f)
    
    model = Model(inputs=[img_input, var_input, age_input], outputs=out, name='DenseNet201_AttentionFusion_Metadata')
    # We return the base_model handle so we can unfreeze it later
    return model, base 

model, base_model = create_densenet_metadata_model()

# --------------------
# == PHASE 1: Train the Head ==
# --------------------
print("\n--- PHASE 1: Training the classification head ---")
checkpoint_path = "densenet_attention_phase1_best.h5"
callbacks_phase1 = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=6, verbose=1, min_lr=1e-7),
    EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True, verbose=1),
    ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1)
]

model.compile(optimizer=Adam(LEARNING_RATE), loss='categorical_crossentropy', metrics=['accuracy'])

# --- Fit Phase 1 ---
history_phase1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_PHASE1,
    callbacks=callbacks_phase1,
    verbose=1
)

# --- Load the best weights from Phase 1 ---
print(f"\nLoading best weights from Phase 1: {checkpoint_path}")
model.load_weights(checkpoint_path)

# --------------------
# == PHASE 2: Fine-Tuning ==
# --------------------
print("\n--- PHASE 2: Fine-tuning the top layers ---")
# Unfreeze the base model
base_model.trainable = True

# --- Freeze all layers except the top 50 ---
layers_to_freeze = len(base_model.layers) - 50 
print(f"Unfreezing top 50 layers of {base_model.name}. Freezing first {layers_to_freeze} layers.")
for layer in base_model.layers[:layers_to_freeze]:
    layer.trainable = False

# --- Re-compile with a very low learning rate ---
model.compile(
    optimizer=Adam(LEARNING_RATE / 10), # e.g., 1e-5
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# --- Create new callbacks for fine-tuning ---
finetune_checkpoint_path = "densenet_attention_finetune_best.h5"
callbacks_finetune = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, verbose=1, min_lr=1e-8), # More sensitive
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1), # Shorter patience
    ModelCheckpoint(finetune_checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1)
]

# --- Fit Phase 2 (Fine-tuning) ---
history_phase2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_PHASE2, # Use new, shorter epoch count
    callbacks=callbacks_finetune,
    verbose=1
)

# Save the final fine-tuned model
model.save("densenet_attention_metadata_finetuned_final.h5")
print("âœ… Saved final fine-tuned model: densenet_attention_metadata_finetuned_final.h5")

# --------------------
# Plot & Save training curves
# --------------------
def save_training_plots(history, prefix="densenet_attention"):
    # Accuracy
    plt.figure(figsize=(10,4))
    plt.plot(history.history['accuracy'], label='train_acc')
    plt.plot(history.history['val_accuracy'], label='val_acc')
    plt.title(f'Accuracy ({prefix})')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.legend()
    plt.grid(alpha=0.3)
    acc_png = f"{prefix}_accuracy.png"
    plt.savefig(acc_png, bbox_inches='tight', dpi=150)
    print(f"Saved {acc_png}")
    plt.close()
    
    # Loss
    plt.figure(figsize=(10,4))
    plt.plot(history.history['loss'], label='train_loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.title(f'Loss ({prefix})')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend()
    plt.grid(alpha=0.3)
    loss_png = f"{prefix}_loss.png"
    plt.savefig(loss_png, bbox_inches='tight', dpi=150)
    print(f"Saved {loss_png}")
    plt.close()

# --- Save plots for both phases ---
print("\nSaving training plots...")
save_training_plots(history_phase1, prefix="densenet_attention_phase1_head")
save_training_plots(history_phase2, prefix="densenet_attention_phase2_finetune")

# --------------------
# Evaluation on test set
# --------------------
# The model object already has the best weights from Phase 2
# (due to EarlyStopping's restore_best_weights=True)
print("\n--- FINAL EVALUATION (after fine-tuning) ---")
test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print(f"\nâœ… Final Test accuracy: {test_acc*100:.2f}% | test loss: {test_loss:.4f}")

# Gather predictions for detailed metrics
y_true = []
y_pred = []

for (imgs, vars_, ages), labs in test_ds:
    preds = model.predict([imgs, vars_, ages], verbose=0)
    y_true.extend(np.argmax(labs.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

# Classification report & confusion matrix
print("\nFinal Classification Report (after fine-tuning):")
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix - DenseNet201 + Attention (Fine-Tuned)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.savefig("densenet_attention_finetuned_confusion_matrix.png", bbox_inches='tight', dpi=150)
print("Saved densenet_attention_finetuned_confusion_matrix.png")
plt.show()

# Weighted metrics
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
print(f"\nFinal Weighted metrics: Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}")

# --------------------
# Print model params
# --------------------
print(f"\nTotal parameters: {model.count_params():,}")
# Count trainable vs. non-trainable
trainable_params = np.sum([np.prod(v.get_shape()) for v in model.trainable_weights])
non_trainable_params = np.sum([np.prod(v.get_shape()) for v in model.non_trainable_weights])
print(f"Trainable params: {trainable_params:,}")
print(f"Non-trainable params: {non_trainable_params:,}")

# Done
print("\nâœ… Done. Fine-tuning complete. Model, plots and confusion matrix saved.")
```

```python
# ================================================================
# DenseNet201 + CBAM (Channel + Spatial Attention) + Metadata (Early Fusion)
# with EarlyStopping, Fine-Tuning, and Training Curves
# ================================================================

import os, warnings
warnings.filterwarnings('ignore')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.layers import (Input, Dense, Dropout, GlobalAveragePooling2D,
                                     GlobalMaxPooling2D, Concatenate, Embedding,
                                     Flatten, BatchNormalization, Multiply, Add,
                                     Activation, Conv2D, Reshape, Lambda)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (EarlyStopping, ReduceLROnPlateau,
                                        ModelCheckpoint)
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import backend as K
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

# ================================================================
# CONFIG
# ================================================================
DATA_PATH = "/kaggle/input/paddy-leaf-small/Small-80"
TRAIN_DIR = os.path.join(DATA_PATH, "train")
TEST_DIR  = os.path.join(DATA_PATH, "test")
METADATA_PATH = "/kaggle/input/small-meatdata/metadata.csv"
PRETRAINED_WEIGHTS = "imagenet"
IMG_SIZE = 224
BATCH_SIZE = 32
SEED = 42
EPOCHS_PHASE1 = 150   # safe because EarlyStopping will stop earlier
LEARNING_RATE = 1e-4

# ================================================================
# LOAD DATA
# ================================================================
df = pd.read_csv(METADATA_PATH)
if not df['image_id'].iloc[0].lower().endswith('.jpg'):
    df['image_id'] = df['image_id'].astype(str) + ".jpg"
df['image_path'] = df.apply(
    lambda r: os.path.join(TRAIN_DIR if r['split']=='train' else TEST_DIR,
                           r['label'], r['image_id']),
    axis=1)
df = df[df['image_path'].apply(os.path.exists)].reset_index(drop=True)
if df.empty:
    raise FileNotFoundError("No image files found. Check paths.")

label_encoder = LabelEncoder()
df['label_enc'] = label_encoder.fit_transform(df['label'])
num_classes = df['label_enc'].nunique()

variety_encoder = LabelEncoder()
df['variety_enc'] = variety_encoder.fit_transform(df['variety'])
num_varieties = df['variety_enc'].nunique()

df['age_norm'] = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min())

train_df_full = df[df['split']=='train']
test_df = df[df['split']=='test']
train_df, val_df = train_test_split(train_df_full, test_size=0.25,
                                    stratify=train_df_full['label_enc'],
                                    random_state=SEED)

print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
print(f"Classes: {num_classes}, Varieties: {num_varieties}")

# ================================================================
# DATA PIPELINE
# ================================================================
def preprocess_image(image, img_size=(IMG_SIZE, IMG_SIZE)):
    image = tf.image.resize(image, img_size)
    return tf.cast(image, tf.float32) / 255.0

def create_dataset(df_local, batch_size=BATCH_SIZE, shuffle=True):
    image_paths = df_local['image_path'].values
    varieties = df_local['variety_enc'].values
    ages = df_local['age_norm'].values
    labels = tf.keras.utils.to_categorical(df_local['label_enc'], num_classes=num_classes)

    ds = tf.data.Dataset.from_tensor_slices((image_paths, varieties, ages, labels))
    if shuffle:
        ds = ds.shuffle(len(df_local), seed=SEED)

    def _load_fn(path, var, age, label):
        img = tf.io.read_file(path)
        img = tf.io.decode_jpeg(img, channels=3)
        img = preprocess_image(img)
        return (img, tf.cast(var, tf.int32), tf.cast(age, tf.float32)), label

    return ds.map(_load_fn, num_parallel_calls=tf.data.AUTOTUNE)\
             .batch(batch_size).prefetch(tf.data.AUTOTUNE)

train_ds = create_dataset(train_df)
val_ds   = create_dataset(val_df, shuffle=False)
test_ds  = create_dataset(test_df, shuffle=False)

# ================================================================
# CBAM BLOCK (fixed with Lambda layers)
# ================================================================
def cbam_block(feature_map, ratio=8, name_prefix="cbam"):
    ch = feature_map.shape[-1]

    # Channel Attention
    shared_dense1 = Dense(ch // ratio, activation='relu', name=f"{name_prefix}_ch_dense1")
    shared_dense2 = Dense(ch, activation=None, name=f"{name_prefix}_ch_dense2")
    avg_pool = shared_dense2(shared_dense1(GlobalAveragePooling2D()(feature_map)))
    max_pool = shared_dense2(shared_dense1(GlobalMaxPooling2D()(feature_map)))
    ch_attn = Activation('sigmoid')(Add()([avg_pool, max_pool]))
    ch_attn = Reshape((1,1,ch))(ch_attn)
    x = Multiply()([feature_map, ch_attn])

    # Spatial Attention
    avg_sp = Lambda(lambda t: K.mean(t, axis=-1, keepdims=True))(x)
    max_sp = Lambda(lambda t: K.max(t, axis=-1, keepdims=True))(x)
    concat = Concatenate(axis=-1)([avg_sp, max_sp])
    sa = Conv2D(1, 7, padding='same', activation='sigmoid', name=f"{name_prefix}_sp_conv")(concat)
    x = Multiply()([x, sa])
    return x

# ================================================================
# BUILD MODEL
# ================================================================
def build_densenet_cbam_metadata(num_classes, num_varieties,
                                 pretrained_weights=PRETRAINED_WEIGHTS,
                                 fine_tune_backbone=False):
    img_input = Input(shape=(IMG_SIZE, IMG_SIZE, 3), name='image_input')
    base = DenseNet201(include_top=False, weights=pretrained_weights,
                       input_tensor=img_input)
    base.trainable = False

    x = cbam_block(base.output, ratio=8, name_prefix="cbam1")
    x = GlobalAveragePooling2D()(x)

    # Image branch
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    image_features = Dense(128, activation='relu')(x)

    # Metadata branch
    var_input = Input(shape=(1,), dtype='int32', name='var_input')
    age_input = Input(shape=(1,), dtype='float32', name='age_input')
    var_emb = Embedding(input_dim=num_varieties, output_dim=8)(var_input)
    var_emb = Flatten()(var_emb)
    var_dense = Dense(32, activation='relu')(var_emb)
    age_dense = Dense(16, activation='relu')(age_input)

    fused = Concatenate()([image_features, var_dense, age_dense])

    f = Dense(256, activation='relu')(fused)
    f = BatchNormalization()(f)
    f = Dropout(0.4)(f)
    f = Dense(128, activation='relu')(f)
    f = BatchNormalization()(f)
    f = Dropout(0.3)(f)
    f = Dense(64, activation='relu')(f)
    f = Dropout(0.2)(f)
    out = Dense(num_classes, activation='softmax')(f)

    model = Model(inputs=[img_input, var_input, age_input], outputs=out)
    if fine_tune_backbone:
        for layer in base.layers[-80:]:
            layer.trainable = True
    return model, base

# ================================================================
# PHASE 1: TRAIN FROZEN BACKBONE
# ================================================================
model, base_model = build_densenet_cbam_metadata(num_classes, num_varieties)
model.compile(optimizer=Adam(LEARNING_RATE),
              loss='categorical_crossentropy', metrics=['accuracy'])

callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=6, verbose=1, min_lr=1e-7),
    EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True, verbose=1),
    ModelCheckpoint("cbam_phase1_best.h5", monitor='val_accuracy', save_best_only=True, verbose=1)
]

history = model.fit(train_ds, validation_data=val_ds,
                    epochs=EPOCHS_PHASE1, callbacks=callbacks, verbose=1)
model.save("cbam_phase1_final.h5")

# ================================================================
# PHASE 2: FINE-TUNE LAST LAYERS
# ================================================================
for layer in base_model.layers[-80:]:
    layer.trainable = True
model.compile(optimizer=Adam(1e-5),
              loss='categorical_crossentropy', metrics=['accuracy'])

ft_callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, verbose=1, min_lr=1e-8),
    EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True, verbose=1),
    ModelCheckpoint("cbam_phase2_best.h5", monitor='val_accuracy', save_best_only=True, verbose=1)
]

history_ft = model.fit(train_ds, validation_data=val_ds,
                       epochs=30, callbacks=ft_callbacks, verbose=1)
model.save("cbam_finetuned_final.h5")

# ================================================================
# TRAINING CURVES (Phase 1 & 2)
# ================================================================
def plot_training_curves(history, title_prefix="Phase1"):
    acc = history.history.get('accuracy', [])
    val_acc = history.history.get('val_accuracy', [])
    loss = history.history.get('loss', [])
    val_loss = history.history.get('val_loss', [])

    # Accuracy
    plt.figure(figsize=(8,5))
    plt.plot(acc, label='Train Accuracy', linewidth=2)
    plt.plot(val_acc, label='Val Accuracy', linewidth=2)
    plt.title(f"{title_prefix} - Accuracy")
    plt.xlabel('Epochs'); plt.ylabel('Accuracy')
    plt.legend(); plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{title_prefix.lower()}_accuracy_curve.png", dpi=150)
    plt.show()

    # Loss
    plt.figure(figsize=(8,5))
    plt.plot(loss, label='Train Loss', linewidth=2)
    plt.plot(val_loss, label='Val Loss', linewidth=2)
    plt.title(f"{title_prefix} - Loss")
    plt.xlabel('Epochs'); plt.ylabel('Loss')
    plt.legend(); plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{title_prefix.lower()}_loss_curve.png", dpi=150)
    plt.show()

    print(f"âœ… Saved plots: {title_prefix.lower()}_accuracy_curve.png / _loss_curve.png")

plot_training_curves(history, title_prefix="Phase1_CBAM")
plot_training_curves(history_ft, title_prefix="Phase2_FineTune_CBAM")

# ================================================================
# EVALUATION
# ================================================================
test_loss, test_acc = model.evaluate(test_ds)
print(f"\nâœ… Test Accuracy: {test_acc*100:.2f}% | Loss: {test_loss:.4f}")

y_true, y_pred = [], []
for (imgs, vars_, ages), labs in test_ds:
    preds = model.predict([imgs, vars_, ages], verbose=0)
    y_true.extend(np.argmax(labs.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix - DenseNet201 + CBAM + Metadata")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout()
plt.savefig("cbam_confusion_matrix.png", dpi=150)
plt.show()

precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
print(f"\nWeighted Precision={precision:.4f}  Recall={recall:.4f}  F1={f1:.4f}")
print(f"\nTotal parameters: {model.count_params():,}")
```
